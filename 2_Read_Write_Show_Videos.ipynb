```python
import cv2
```

```python
# setting the variable to capture the video
cap = cv2.VideoCapture(0) # 0 is camera number in your system, if it doesn't work try -1,1,2. We can also read an existing video file by replacing 0 with the file name.
```

```python
# setting the video codec
# to know more about video codec:
# https://www.fourcc.org/fourcc.php
fourcc = cv2.VideoWriter_fourcc(*'XVID')
```

```python
# setting the variable to save the video
out = cv2.VideoWriter('capture_1.avi', fourcc, 30.0, (640,480)) # file_name, video codec, frame rate, size(width, height)
```

```python
# capturing the video
while(cap.isOpened()):
  ret, frame = cap.read() # ret: will hold true or false depending on wheather there is video source present or not.frame: will capture the video
  if ret == True:
      # viewing various properties of the captured video
      # list of properties that you can see: 
      # https://docs.opencv.org/4.0.0/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d
      print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
      print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

      # saving the captured video
      out.write(frame)

      # to get the video is gray scale. by default the video is capture in BGR
      gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
      # displaying the captured video
      cv2.imshow('capture_1', gray)
      # We will use the above command to display but because of defect in cobal we are using this
      # cv2_imshow(frame)

      if cv2.waitKey(1) & 0xFF == ord('q'): # this will wait capture the video till the user presses 'q' button on the keyboard
        break
  else:
    break

cap.release() # will release the resources 
out.release() # will release the resources 
cv2.destroyAllWindows() # will close all the windows created by cv2
```
